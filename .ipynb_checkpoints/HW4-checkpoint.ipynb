## b.1) Wait And See (WAS) model

```julia
function solve_WAS(s::Int64; demand = demand, transportation = transportation, production = production, zones = zones, facilities = facilities)
    modelWAS = Model(Gurobi.Optimizer)  # Builds Gurobi model
    
    # Define the variables
    @variable(modelWAS, xj[1:20] >= 0, Bin)  # decision to build a facility j or not (Binary)
    @variable(modelWAS, yij[1:50, 1:20] >= 0)  # amount of product sent from facility `j` to demand point `i`
    @variable(modelWAS, zi[1:50] >= 0)  # unmet demand for each demand point `i`

    # Define the objective
    @objective(modelWAS, Min, 
        facilities[:,1]' * xj +                # facility setup cost
        sum(transportation .* yij) +           # transportation cost
        sum(zones .* zi) +                     # unmet demand cost
        sum(production[j,s] * sum(yij[i,j] for i = 1:50) for j = 1:20)  # production cost
    )

    # Demand constraint: Ensure demand or unmet demand is met
    @constraint(modelWAS, c1[i=1:50], sum(yij[i,j] for j =1:20) + zi[i] .≥ demand[i,s])

    # Production capacity constraint for each facility
    @constraint(modelWAS, c2[j=1:20], sum(yij[i,j] for i =1:50) .≤ xj[j] * facilities[j,2])

    # Solve the model
    optimize!(modelWAS)
    
    # Check if the model solved successfully
    if termination_status(modelWAS) == MOI.OPTIMAL
        println("$s Done!")
        return value.(xj), value.(yij),  value.(zi), objective_value(modelWAS)
    else
        @error "$s fails"
    end
end
```

```julia
function solve_WAS2(s::Int64; demand = demand, transportation = transportation, production = production, zones = zones, facilities = facilities)
    modelWAS = Model(Gurobi.Optimizer)  # Builds Gurobi model
    
    # Define the variables
    @variable(modelWAS, xj[1:20] >= 0, Bin)  # decision to build a facility j or not (Binary)
    @variable(modelWAS, yij[1:50, 1:20] >= 0)  # amount of product sent from facility `j` to demand point `i`
    @variable(modelWAS, zi[1:50] >= 0)  # unmet demand for each demand point `i`

    # Define the objective
    @objective(modelWAS, Min, 
        facilities[:,1]' * xj +                # facility setup cost
        sum(transportation .* yij) +           # transportation cost
        sum(zones .* zi) +                     # unmet demand cost
        sum(production[:,s] * sum(yij, dims = 1) )  # production cost
    )

    # Demand constraint: Ensure demand or unmet demand is met
    @constraint(modelWAS, c1, sum(yij, dims = 2) + zi .≥ demand[:,s])

    # Production capacity constraint for each facility
    @constraint(modelWAS, c2, sum(yij, dims = 1)' .≤ xj .* facilities[:,2])

    # Solve the model
    optimize!(modelWAS)
    
    # Check if the model solved successfully
    if termination_status(modelWAS) == MOI.OPTIMAL
        println("$s Done!")
        return value.(xj), value.(yij),  value.(zi), objective_value(modelWAS)
    else
        @error "$s fails"
    end
end
```

```julia
production[:,1] * sum(yij, dims = 1) 
```

```julia
#Note: After optimization, sum the objective value and times 0.002 (Expected value).
set_x = Vector(undef, 500)
set_y = Vector{Matrix}(undef, 500)
set_z = Vector(undef, 500)
set_objectives = zeros(500,1)

for s = 1:500
    @show s
    x , y, z, objective = solve_WAS2(s)
    set_x[s] = x
    set_y[s] = y
    set_z[s] = z
    @show set_objectives[s] = objective
end
    
```

```julia
objWAS = 0.002*sum(set_objectives)
```

## b.2) Adaptive Stochastic Optimization (ASO) model

```julia
modelASO = Model(Gurobi.Optimizer); # Builds Gurobi model

# Define the variables
@variable(modelASO, xj[1:20] ≥ 0, Bin); #decision to build a facility j or not (Binary).
@variable(modelASO, yijs[1:50,1:20,1:500] ≥ 0) ; #amount of product sent from facility `j` to demand point `i` for a scenario `s`.
@variable(modelASO, zis[1:50, 1:500] ≥ 0) ; #unmet demand on demand i scenario s.

# # Define the objective
@objective(modelASO, Min, facilities[:,1]'*xj + #facility setup cost
                0.002*(
                sum(transportation[i,j]*yijs[i,j,s] for i =1:50 for j =1:20 for s = 1:500) +  #transportation cost
                sum(zones.*sum(zis, dims = 2)) +              #unmet demand cost
                sum(production.*sum(yijs, dims = 1)[1,:,:]))      #production cost
)
# Demand constraint 
@constraint(modelASO, c1[i=1:50, s = 1:500] , sum(yijs[i,:,s]) +  zis[i,s] ≥ demand[i,s])

#production capacity constraint
@constraint(modelASO, c2[s=1:500], sum(yijs[:,:,s], dims = 1)' .≤ xj.*facilities[:,2])
```

```julia
# Optimize model
optimize!(modelASO);
@show is_solved_and_feasible(modelASO);
@show objective_value(modelASO);
```

## b.3) Deterministic (DTM) Model

```julia
modelDTM = Model(Gurobi.Optimizer);

# Define the variables
@variable(modelDTM, xj[1:20] ≥ 0, Bin); #decision to build a facility j or not (Binary).
@variable(modelDTM, yij[1:50,1:20] ≥ 0) ; #amount of product sent from facility `j` to demand point `i`
@variable(modelDTM, zis[1:50,1:500] ≥ 0) ; #unmet demand on demand i scenario s.

# # Define the objective
@objective(modelDTM, Min, facilities[:,1]'*xj +  #facility setup cost
                0.002*(sum(transportation.*yij)+  #transportation cost
                sum(zones.*sum(zis, dims = 2)) +   #unmet demand cost
                sum(sum(production, dims = 2)'.*sum(yij, dims=1))  #production cost
                )
)

# Demand constraint 
@constraint(modelDTM, c1[i=1:50, s = 1:500], sum(yij[i,:]) + zis[i,s] ≥ demand[i,s])

#production capacity constraint
@constraint(modelDTM, c2[s=1:500], sum(yij, dims = 1)' .≤ xj.*facilities[:,2])
```

```julia
# Optimize model
optimize!(modelDTM);
@show is_solved_and_feasible(modelDTM);
@show objective_value(modelDTM);
```

```julia
function vis_result(y::Array{VariableRef, N}; demand = demand, title = "Demand Fulfilment") where N
    dictDemand = Dict( s => sum(demand[:,s]) for s = 1:500)
    sortScenario = sort(collect(dictDemand), by = x->x[2])
    vals = [x[2] for x in sortScenario]
    idx = [x[1] for x in sortScenario]
    y = value.(y)
    if size(y,3) != 1 #
        @show title = "ASO"
        dictY = Dict( s => sum(y[:,:,s]) for s = 1:500) 
    else 
        @show title = "DTM"
        dictY = Dict( s => sum(y) for s = 1:500) 
    end
        yvals = [dictY[id] for id in idx]


    f1 = Figure(size = (500,500))
    ax1 = Axis(f1[1,1], title = title) 


    barplot!(ax1, 1:500, vals)

    barplot!(ax1, 1:500, yvals)
    return f1
end

function vis_result(set_y::Vector{Matrix}; demand = demand, title = "Demand Fulfilment")
    f1 = Figure(size = (500,500))
    ax1 = Axis(f1[1,1], title = "WAS") 
    dictDemand = Dict( s => sum(demand[:,s]) for s = 1:500)
    sortScenario = sort(collect(dictDemand), by = x->x[2])
    vals = [x[2] for x in sortScenario]
    idx = [x[1] for x in sortScenario]

    barplot!(ax1, 1:500, vals)

    total_y = zeros(500,1)

    for s in 1:500
        # @show y[s][:yijs]
        total_y[s]  = sum(set_y[s])
    end

    dictY = Dict( s => total_y[s] for s = 1:500) 

    yvals = [dictY[id] for id in idx]

    barplot!(ax1, 1:500, yvals)
    return f1
end
```

```julia
D = (CSV.File("HW4_data/demand.csv",header=1) |> Tables.matrix)[:,2:end]; # demand (i,s)
c = (CSV.File("HW4_data/facilities.csv",header=1) |> Tables.matrix)[:,2]; # construction cost (j)
C = (CSV.File("HW4_data/facilities.csv",header=1) |> Tables.matrix)[:,3]; # capacity (j)
q = (CSV.File("HW4_data/production.csv",header=1) |> Tables.matrix)[:,2:end]; # production cost (j,s)
t = (CSV.File("HW4_data/transportation.csv",header=1) |> Tables.matrix)[:,2:end]; # transportaton cost (i,j)
u = (CSV.File("HW4_data/zones.csv",header=1) |> Tables.matrix)[:,2:end]; # cost of unmet demand (i)
```

```julia
n = size(c)[1] # number of facilities
m = size(u)[1] # number of customer zones
S = size(q)[2] # number of scenarios
println("Number of facilities: ", n)
println("Number of customers zones: ", m)
println("Number of scenarios: ", S)
# probability vector
p = 1/S * ones(S)
println("Probability of each scenario: ", p[1])
```

```julia
model = Model(() -> Gurobi.Optimizer(GRB_ENV))
@variable(model, x[1:n], Bin); # whether or not to build a facility
@variable(model, y[1:m, 1:n, 1:S] >= 0) # quantity produced and transported from facility j to customer i in scenario s
@variable(model, z[1:m, 1:S] >= 0); # unmet demand from customer i in scenario s
@constraint(model, [j in 1:n, s in 1:S], sum(y[i, j, s] for i in 1:m) <= C[j] * x[j]) # remain under capacity
@constraint(model, [i in 1:n, s in 1:S], z[i, s] == D[i, s] - sum(y[i, j, s] for j in 1:n)); # unmet demand

@objective(model, Min,
sum(c[j] * x[j] for j in 1:n) +
sum(p[s] * t[i, j] * y[i, j, s] for j in 1:n, i in 1:m, s in 1:S) +
sum(p[s] * u[i] * z[i, s] for i in 1:m, s in 1:S) +
sum(p[s] * q[j, s] * sum(y[i, j, s] for i in 1:m) for j in 1:n, s in 1:S));

optimize!(model)
facility_SO = value.(x);
transport_SO = value.(y);
unmet_SO = value.(z);
exp_cost_SO = objective_value(model)
println("Number of facilities constructed: ", sum(facility_SO))
println("Stochastic cost: ", exp_cost_SO)
```

```julia
model = Model(() -> Gurobi.Optimizer(GRB_ENV))
@variable(model, x[1:n], Bin)
@variable(model, y[1:m, 1:n] >= 0)
@variable(model, z[1:m] >= 0)
@constraint(model, [j in 1:n], sum(y[i, j] for i in 1:m) <= C[j] * x[j])
@constraint(model, [i in 1:n], z[i] == sum(D[i, :])/50 - sum(y[i, j] for j in 1:n));
@objective(model, Min,
sum(c[j] * x[j] for j in 1:n) +
sum(t[i, j] * y[i, j] for j in 1:n, i in 1:m) +
sum(u[i] * z[i] for i in 1:m) +
sum(q[j] * sum(y[i, j] for i in 1:m) for j in 1:n))
optimize!(model)
facility_EV = value.(x)
println("Number of facilities constructed: ", sum(facility_EV))
println("Cost: ", objective_value(model))
```

```julia
cost_scenario = zeros(S);
transport_EV = zeros(m,n,S);
unmet_EV = zeros(m,S);
production_EV = zeros(n,S);
for s in 1:S
model = Model(() -> Gurobi.Optimizer(GRB_ENV))
@variable(model, y[1:m,1:n] >= 0)
@variable(model, z[1:m] >= 0)
@constraint(model, [j in 1:n], sum(y[i, j] for i in 1:m) <= C[j] * facility_EV[j])
@constraint(model, [i in 1:n], z[i] == D[i, s] - sum(y[i, j] for j in 1:n));
@objective(model, Min,
sum(c[j] * facility_EV[j] for j in 1:n) +
sum(t[i, j] * y[i, j] for j in 1:n, i in 1:m) +
sum(u[i] * z[i] for i in 1:m) +
sum(q[j] * sum(y[i, j] for i in 1:m) for j in 1:n))
optimize!(model)
cost_scenario[s] = objective_value(model);
transport_EV[:,:,s] = value.(y);
unmet_EV[:,s] = value.(z);
end
exp_cost_EV = sum(p[s] * cost_scenario[s] for s in 1:S)
println("Deterministic cost: ", exp_cost_EV)
```
