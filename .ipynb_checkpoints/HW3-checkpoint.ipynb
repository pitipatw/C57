```julia
modelA = Model(Gurobi.Optimizer); # Builds Gurobi model

# distance between center i and landfill j
d = [norm(centers[i,:] - landfills[j,:]) for i in eachindex(centers[:,1]), j in eachindex(landfills[:,1])]

# Define the variables
@variable(modelA, x[1:50,1:15] ≥ 0); #decision to send waste from station i to landfill j or not?
@variable(modelA, y[1:15] ≥ 0 , Bin) ; # decision to build the landfill or not?

#somehow, I think it should be this ... 
# @objective(modelA, Min, sum(d*q)); # cij = distance_ij * WasteWeight i * station i * xj

# Define the objective
# @objective(modelA, Min, sum(q'*d.*x));
@objective(modelA, Min, sum( [ q[i]*d[i,j]*x[i,j] for i =1:50 for j = 1:15]))

# Build no more than 5 landfills
@constraint(modelA, c1 , sum(y) == 5);

# Make sure every stations has 1, and only 1 ,landfill to go to
@constraint(modelA, c2[i = 1:50] , sum(x[i,:]) == 1) 

# Make sure that each landfill that built also serve those stations
@constraint(modelA, c3[i=1:50, j=1:15], x[i,j] ≤ y[j])

optimize!(modelA); # Optimize model
@show is_solved_and_feasible(modelA);
```

```julia
modelB = Model(Gurobi.Optimizer); # Builds Gurobi model

# distance between center i and landfill j
d1 = [norm(centers[i,:] - landfills[j,:]) for i in eachindex(centers[:,1]), j in eachindex(landfills[:,1])]
d2 = [norm(stations[i,:] - centers[j,:]) for i in eachindex(stations[:,1]), j in eachindex(centers[:,1])]
d3 = [norm(stations[i,:] - landfills[j,:]) for i in eachindex(stations[:,1]), j in eachindex(landfills[:,1])]

# Define the variables
@variable(modelB, y[1:15] ≥ 0 , Bin) ; # decision to build the landfill or not?
@variable(modelB, z[1:50] ≥ 0 , Bin) ; # decision to build the station or not?
@variable(modelB, w[1:50,1:50] ≥ 0); 
@variable(modelB, v[1:50,1:15] ≥ 0); 
@variable(modelB, x[1:50,1:15] ≥ 0); 

# Define the objective
# @objective(modelA, Min, sum(q'*d.*x));
@objective(modelB, Min, sum( q[i]*x[i,j]*d1[i,j] for i =1:50 for j = 1:15) +
                sum( q[i]*w[i,k]*d2[i,k] for i =1:50 for k = 1:50) +
            0.5*sum( d3[k,j]* v[k,j]*sum(q[i]*w[i,k] for i = 1:50) for k = 1:50 for j = 1:15)+
            sum(10000*z) );                    

# Build no more than 5 landfills
@constraint(modelB, c1 , sum(y) == 5);

@constraint(modelB, c2[i = 1:50], sum(x[i,:]) + sum(w[i,:]) == 1)

@constraint(modelB, c3[k = 1:50], sum(q'*w[:,k]) ≤ 2000*z[k]) 

# Make sure every stations has 1, and only 1 ,landfill to go to
@constraint(modelB, c4[i=1:50, j=1:15], x[i,j] ≤ y[j])

# in = out
# Make sure that each landfill that built also serve those stations
@constraint(modelB, c5[k=1:50, j=1:15], v[k,j] ≤ z[k]*y[j]*1000000)

optimize!(modelB); # Optimize model
@show is_solved_and_feasible(modelB);
```
